```python
import pandas as pd
pd.set_option('display.max_columns',100)
```

## Importing data

```python
# Define the paths to the files
apparel_path = '../data/extracted/amazon_reviews_us_Apparel_v1_00.tsv'
beauty_path = '../data/extracted/amazon_reviews_us_Beauty_v1_00.tsv'
shoes_path = '../data/extracted/amazon_reviews_us_Shoes_v1_00.tsv'

# Read the files into dataframes
df_apparel_reviews = pd.read_csv(apparel_path, sep='\t', on_bad_lines='skip')
df_beauty_reviews = pd.read_csv(beauty_path, sep='\t', on_bad_lines='skip')
df_shoes_reviews = pd.read_csv(shoes_path, sep='\t', on_bad_lines='skip')
```

```python
print("Apparel reviews dimensions:",df_apparel_reviews.shape )
display(df_apparel_reviews.head(5))
print("\nBeauty reviews dimensions:",df_beauty_reviews.shape )
display(df_beauty_reviews.head(5))
print("\nShoes reviews dimensions:",df_shoes_reviews.shape )
display(df_shoes_reviews.head(5))
```

```python
# check if all three dataframes have same column names

if all(df_apparel_reviews.columns == df_beauty_reviews.columns) &  all(df_apparel_reviews.columns == df_shoes_reviews.columns) :
    print("Columns same")
else:
    print("Check dataframes, columns not same")
```

```python
df_merged= pd.concat([df_apparel_reviews,df_shoes_reviews ])
```
